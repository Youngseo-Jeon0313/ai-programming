```python
import matplotlib.pyplot as plt
import numpy as np
```

  (1) plt.plot으로 임의의 그래프 1개를 만드세요

```python
x1=[1,2,3]
y1=[2,4,6]
plt.plot(x1,y1)
plt.title("My Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

(2) plt.bar으로 임의의 그래프 1개를 만드세요

```python
x2=np.arange(-5,8)
y2=x2**2
plt.bar(x2,y2)
plt.show()
```

  (3) plt.scatter으로 임의의 그래프 1개를 만드세요

```python
x3 = np.random.rand(10)
y3 = np.random.rand(10)
colors = np.random.randint(0,100,10)
sizes = np.pi * 1000 * np.random.rand(10)
plt.scatter(x3,y3,c=colors, s=sizes, alpha=0.7)
plt.show()
```

  (4) plt.hist 으로 임의의 그래프 1개를 만드세요


```python
data = np.random.randn(1000)
f,(ax1, ax2) = plt.subplots(1,2,figsize = (6,3)) 
#2,1로 하면 세로로 보여줌. 불안정하면 figsize를 6,12로 조정하면 됨!
#histogram (pdf)
ax1.hist(data, bins=30, density=True, color='b')
#empirical cdf
ax2.hist(data, bins=30, density=True, color='r', cumulative=True)
```

(5) plt.subplots 를 이용해서 위 그래프 중 2개를 1개의 figure에 넣어보세요


```python
fig,axes=plt.subplots(2,1) #2개의 그래프가 들어가는 figure 생성
axes[0].plot(x1, y1)
axes[1].plot(x2, y2)
#하나의 이미지안에 여러 가지의 plot을 보여주고자 할 때
```

  (7) 임의의 함수를 정하고 이를 미분한 함수를 같이 만드세요. 이 둘을 한개의 그래프에 같이 표시하고 미분값이 의미하는 바를 적어보세요.

```python
x7 = np.linspace(0,10,50)
y7_1 = x7**3 + 6*x7 + 3
y7_2 =3*x7**2 +6

plt.plot(x7,y7_1,linestyle="-",color="red",label="$x^3 + 6x + 3$")
plt.plot(x7,y7_2,linestyle=":",color="blue",label="$3x^2 + 6$")
plt.legend(shadow = True, borderpad = 2)
plt.show()
```
